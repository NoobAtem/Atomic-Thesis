```python
# Let's create a checkpoint based on time
checkpoint: str = None
def create_checkpoint(name: str) -> None:
    if not os.path.exists(os.path.join(MPATH, PREF + MEFF)):
        os.mkdir(os.path.join(MPATH, PREF + MEFF))    
    checkpoint = os.path.join(MPATH, PREF + MEFF)

def generate_data() -> None:
    if not os.path.exists(os.path.join(checkpoint, "train")):
        os.mkdir(os.path.join(checkpoint, "train"))
    if not os.path.exists(os.path.join(checkpoint, "val")):
        os.mkdir(os.path.join(checkpoint, "val"))    
    if not os.path.exists(os.path.join(checkpoint, "test")):
        os.mkdir(os.path.join(checkpoint, "test"))    

def generate_set(train_size=0.8, val_size=0.1, test_size=0.1):
    if train_size + val_size + test_size != 100:
        print("Size is not uniformed")
        return
    pig_data: list = os.listdir(DPPPATH)
    bat_data: list = os.listdir(DPBPATH)
    frog_data: list = os.listdir(DPFPATH)
    
    pig_count: int = len(pig_data)
    bat_count: int = len(bat_data)
    frog_count: int = len(frog_data)

    pig_trainsize: int = int(pig_count*train_size)
    bat_trainsize: int = int(bat_count_*train_size)
    frog_trainsize: int = int(frog_count*train_size)

    pig_valsize: int = int(pig_count*val_size)
    bat_valsize: int = int(bat_count_*val_size)
    frog_valsize: int = int(frog_count*val_size)

    pig_testsize: int = int(pig_count*test_size)
    bat_testsize: int = int(bat_count_*test_size)
    frog_testsize: int = int(frog_count*test_size)

    pig_start: int = rd.randint(0, pig_count - pig_trainsize)
    bat_start: int = rd.randint(0, bat_count - bat_trainsize)
    frog_start: int = rd.randint(0, frog_count - frog_trainsize)

    pig_end: int = pig_start + pig_trainsize
    bat_end: int = bat_start + bat_trainsize
    frog_end: int = frog_start + frog_trainsize

    for spec in pig_data[pig_start:pig_end]:
        shutil.copy(os.path.join(DPPPATH, spec), os.path.join(MPATH, "train", spec))
        
    del pig_data[pig_start:pig_end]
    pig_start: int = rd.randint(0, pig_count - pig_trainsize)
    bat_start: int = rd.randint(0, bat_count - bat_trainsize)
    frog_start: int = rd.randint(0, frog_count - frog_trainsize)
    for spec in pig_data[pig_start:pig_end]:
        shutil.copy(os.path.join(DPPPATH, spec), os.path.join(MPATH, "train", spec))
    for spec in pig_data[pig_start:pig_end]:
        shutil.copy(os.path.join(DPPPATH, spec), os.path.join(MPATH, "train", spec))
```
